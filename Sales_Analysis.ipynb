```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("archive.zip") # load dataset
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')  # date conversion 
```

```python
df = df.dropna(subset=['Order Date'])   # invalid dates remove
```

```python
df['Month'] = df['Order Date'].dt.month     # month column
```

```python
monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales.plot(kind='line')

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")

plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()
category_sales.plot(kind='bar')

plt.xlabel("Category")
plt.ylabel("Sales")
plt.title("Category-wise Sales")

plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum()

region_sales.plot(kind='bar')

plt.xlabel("Region")
plt.ylabel("Sales")
plt.title("Region-wise Sales")

plt.show()
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)


top_products.plot(kind='bar')

plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Top 10 Products")

plt.show()
```

```python
segment_sales = df.groupby('Segment')['Sales'].sum()

segment_sales.plot(kind='pie', autopct='%1.1f%%')

#plt.xlabel("Region")
#plt.ylabel("Sales")
plt.title("Sales by Segment")

plt.show()
```

```python
ship_mode = df['Ship Mode'].value_counts()

ship_mode.plot(kind='bar')

#plt.xlabel("Region")
#plt.ylabel("Sales")
plt.title("shipping mode usage")

plt.show()
```

# BUSINESS INSIGHTS

MONTHLY SALES TREND - Sales peaked during the end months of the year, indicating possible seasonal demand.

CATEGORY-WISE-SALES - Technology category generated the highest overall sales revenue despite lower order frequency.

REGION-WISE SALES - West region contributed the highest sales among all regions.

TOP 10 PRODUCTS - A small number of products contributed significantly to total sales revenue.

SALES BY SEGMENT - Consumer segment contributed the highest share of overall sales, indicating strong demand from individual customers.

SHIP MODE ANALYSIS - Standard Class was the most frequently used shipping mode, suggesting customer preference for cost-effective delivery.

# CONCLUSION 

The analysis revealed that Technology products generated the highest sales revenue, while the West region contributed the most to overall sales. Seasonal trends indicated higher sales toward the end of the year.
